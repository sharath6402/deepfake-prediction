```python
import cv2
import os

def convert_videos_to_images(video_dir, image_dir):
    # Create the output directory if it doesn't exist
    if not os.path.exists(image_dir):
        os.makedirs(image_dir)

    # Iterate through each video file in the directory
    for video_file in os.listdir(video_dir):
        if video_file.endswith('.mp4') or video_file.endswith('.avi'):
            video_path = os.path.join(video_dir, video_file)
            video_name = os.path.splitext(video_file)[0]

            # Open the video file
            cap = cv2.VideoCapture(video_path)
            frame_count = 0

            # Read frames until the end of the video
            while cap.isOpened():
                ret, frame = cap.read()

                if ret:
                    # Save the frame as an image
                    image_name = f"{video_name}_frame_{frame_count}.jpg"
                    image_path = os.path.join(image_dir, image_name)
                    cv2.imwrite(image_path, frame)
                    frame_count += 1
                else:
                    break

            # Release the video capture object
            cap.release()
            cv2.destroyAllWindows()

# Specify the directory containing the videos
video_directory = '/content/drive/MyDrive/test'

# Specify the directory where the images will be saved
image_directory = '/content/drive/MyDrive/test_images'

# Convert videos to images
convert_videos_to_images(video_directory, image_directory)

```

```python
import cv2
import os

def resize_images_in_directory(image_dir, size=(256, 256)):
    # Iterate through each image file in the directory
    for image_file in os.listdir(image_dir):
        if image_file.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
            image_path = os.path.join(image_dir, image_file)

            # Read the image
            image = cv2.imread(image_path)

            # Resize the image to the desired size
            resized_image = cv2.resize(image, size)

            # Save the resized image, overwriting the original image
            cv2.imwrite(image_path, resized_image)


# Resize all images in the directory to 256x256
resize_images_in_directory('/content/drive/MyDrive/test_images', size=(256, 256))
```
