```python
import shutil
import os
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/train_data.csv")
X = df.iloc[:, 0].values
y = df.iloc[:, 1].values
source_dir = '/content/drive/MyDrive/dataset/'

for i,j in zip(X, y):
    try:
        if j == 0:

          destination_dir = '/content/drive/MyDrive/datasets/train/real/'
          source_file = os.path.join(source_dir, i)
          destination_file=os.path.join(destination_dir,i)
          if os.path.exists(destination_file):
            print(f"Skipping {i} as it's already resized.")
            continue
          else:
            shutil.copy(source_file, destination_file)
            print(f"File '{i}' copied successfully.")
        elif j==1:
          destination_dir = '/content/drive/MyDrive/datasets/train/fake/'
          source_file = os.path.join(source_dir, i)
          destination_file=os.path.join(destination_dir,i)
          if os.path.exists(destination_file):
            print(f"Skipping {i} as it's already resized.")
            continue
          else:
            shutil.copy(source_file, destination_file)
            print(f"File '{i}' copied successfully.")
        else:
            print(f"Skipping file '{i}'.")
    except Exception as e:
        print(f"Error copying file '{i}': {e}")
print("Train data completely shifted.")
```
