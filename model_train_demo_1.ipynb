```python
import os
import cv2

def resize_videos(input_folder, output_folder, new_size):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # List all video files in the input folder
    video_files = [f for f in os.listdir(input_folder) if f.endswith('.mp4')]

    for video_file in video_files:
        input_video_path = os.path.join(input_folder, video_file)
        output_video_path = os.path.join(output_folder, video_file)

        # Check if the output video file already exists
        if os.path.exists(output_video_path):
            print(f"Skipping {video_file} as it's already resized.")
            continue

        resize_video(input_video_path, output_video_path, new_size)

def resize_video(video_path, output_path, new_size):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Get the original frame size
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Create a VideoWriter object to write the resized video
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec to use
    out = cv2.VideoWriter(output_path, fourcc, cap.get(cv2.CAP_PROP_FPS), new_size)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Resize the frame to the new size
        resized_frame = cv2.resize(frame, new_size)

        # Write the resized frame to the output video
        out.write(resized_frame)

    # Release resources
    cap.release()
    out.release()

# Example usage
input_folder = '/content/drive/MyDrive/fake'
output_folder = '/content/drive/MyDrive/dataset'
new_size = (224, 224)  # New size for the frames
resize_videos(input_folder, output_folder, new_size)
```
